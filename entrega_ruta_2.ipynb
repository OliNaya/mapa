#### Ruta senderismo (ca ruta)

```python
#sacada de https://es.wikiloc.com/
#usamos gpx para leer os archivos gpz
with open('rutas_senderismo/ezaro.gpx', 'r') as gpx_file:
    ezaro = gpxpy.parse(gpx_file)
with open('rutas_senderismo/fontedacoba-cantera-pizarra-a-teixadal.gpx', 'r') as gpx_file:
    fontedacoba = gpxpy.parse(gpx_file)
with open('rutas_senderismo/costa-das-rias-baixas-etapa-13-san-vicente-do-mar-porto-pedr.gpx', 'r') as gpx_file:
    pedras = gpxpy.parse(gpx_file)
with open('rutas_senderismo/subida-al-mustallar-lugo.gpx', 'r') as gpx_file:
    mustallar = gpxpy.parse(gpx_file)
with open('rutas_senderismo/praia-de-area-negra-y-fervenza-en-teixidelo-san-andres-de-te.gpx', 'r') as gpx_file:
    sanandres = gpxpy.parse(gpx_file)
with open('rutas_senderismo/ruta-das-fragas-de-catasos-lalin.gpx', 'r') as gpx_file:
    catasos = gpxpy.parse(gpx_file)
with open('rutas_senderismo/mao-ourense.gpx', 'r') as gpx_file:
    mao = gpxpy.parse(gpx_file)
```

```python
#creo unha función para leelos e almacenos nunha lista unha por ruta
def crear(gpx):
    route = []
    for track in gpx.tracks:
        for segment in track.segments:
            for point in segment.points:
                route.append((
                    
                    point.latitude, point.longitude)
                )
    return route

ezaro_list=crear(ezaro)
fondedacoba_list=crear(fontedacoba)
pedras_list=crear(pedras)
mustallar_list=crear(mustallar)
sanandres_list=crear(sanandres)
catasos_list=crear(catasos)
maho_list=crear(mao)
```

### Mapa

```python


#m = folium.Map(location=[43, -8.20],zoom_start=8, width=1000, height=1000)
m = folium.Map(location=[43, -8.10],zoom_start=-1,width=1000,height=1000,top=5, left=10) #SE A QUERO GARDAR HTML ASI SE A QUERO VER A LINEA DE ARRIBA
# capa tiles
# https://fontawesome.com/v4/icons/ estas iconas funcionan
#
folium.TileLayer('openstreetmap').add_to(m)
folium.TileLayer('Stamen Toner').add_to(m)
folium.TileLayer('CartoDB Positron').add_to(m)
folium.TileLayer('CartoDB dark_matter').add_to(m)
folium.TileLayer('stamenterrain').add_to(m)
folium.TileLayer('stamen watercolor').add_to(m)


# capa comarcas
folium.GeoJson(provincias, name='Provincias').add_to(m)

# clúster
marker_cluster = folium.plugins.MarkerCluster(name='Festas-Restaurantes').add_to(m)
marker_cluster2=folium.plugins.MarkerCluster(name='Casas Rurales').add_to(m)
marker_cluster3=folium.plugins.MarkerCluster(name='Rutas de Senderismo- Rutas as Fervenzas ', show=False).add_to(m) #asi consigo que non apareza por defecto selecionadada
#capas rutas de senderismo
#con popyline non podemos poñer icona
folium.PolyLine(ezaro_list,color='red', weight=4.5,popup=('Ezaro-MontePindo')).add_to(marker_cluster3)
folium.PolyLine(fondedacoba_list,color='orange', weight=4.5,popup=('Fonte da Coba- Pico dos Formigueiros')).add_to(marker_cluster3)
folium.PolyLine(pedras_list,color='violet', weight=4.5,popup=('Pedras Negras')).add_to(marker_cluster3)
folium.PolyLine(mustallar_list,color='pink', weight=4.5,popup=('Subida o Mustallar')).add_to(marker_cluster3)
folium.PolyLine(sanandres_list,color='yellow', weight=4.5,popup=('Praia da Area Negra - Fervenza do teixidelo')).add_to(marker_cluster3)
folium.PolyLine(catasos_list,color='black', weight=4.5,popup=('Ruta das fragas de Catasos')).add_to(marker_cluster3)
folium.PolyLine(maho_list,color='blue', weight=4.5,popup=('Ruta do Cañon do rio Mao')).add_to(marker_cluster3)
# capa festas
for idx, sitio in datos_df.iterrows():  
    try:
        a_a="<a href=\"" + sitio['enlace']+ "\"target='_blank'>" + sitio['nome']+ "</a> "

        folium.Marker([sitio.lon,sitio.lat],popup= a_a,icon=folium.Icon(color='blue',icon='glyphicon-music')).add_to(marker_cluster)

    except:
        pass
# capa casas rurales
for idx, s in datos_st.iterrows(): 
    try:   
        a="<a href=\"" + s['enlace']+ "\"target='_blank'>" + s['nome']+ "</a> "
        folium.Marker([s.lon,s.lat],popup= a ,icon=folium.Icon(color='green',icon='home')).add_to(marker_cluster2)
    except:
        pass
    
# capa restaurantes as nosas propostas metina noutro cluster, por se solo se quere ir a comer

for idx, restaurante in restaurantes_total.iterrows():
    try:
        folium.Marker([restaurante.latitude,restaurante.lonxitude],
        tooltip=restaurante.nome,
        popup=(f'{restaurante.nome} ({restaurante.comentario})'),
        icon=folium.Icon(color='red',icon='beer', prefix='fa')).add_to(marker_cluster)
    except:
        pass
# fervenzas 
for idx, s in datos_rt.iterrows(): 
    try:   
        a="<a href=\"" + s['enlace']+ "\"target='_blank'>" + s['nome']+ "</a> "
        folium.Marker([s.lon,s.lat],popup= a ,icon=folium.Icon(color='darkgreen',icon='fa-map-signs', prefix='fa')).add_to(marker_cluster3)
    except:
        pass

sw = datos_df[['lon', 'lat']].min().values.tolist()
ne = datos_df[['lon', 'lat']].max().values.tolist()

m.fit_bounds([sw, ne])      
    

folium.LayerControl().add_to(m)
#m

m.save('mapa_rutas.html') #asi gardamolo a html
```
